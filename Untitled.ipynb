```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import datetime
from light_sim import LightSim

start = datetime.datetime(2012,1,1)
end = datetime.datetime(2012,1,2)
```

```python

wavelengths= np.arange(300, 800, 1)
# wavelengths=[420, 530, 660 ]
l = LightSim(start, end, latitude=-35.278454, longitude=149.121483, wavelengths=wavelengths)
```

```python
matplotlib.rcParams['figure.figsize'] = (17.0, 20.0)
start_doy = start.timetuple().tm_yday
end_doy = end.timetuple().tm_yday
# resample = 100
# xs = np.arange(start_doy,end_doy)+0.5

xs = np.linspace(1,2, num=1440)

# xs = np.linspace(start_doy, end_doy, num=60*1440)

cmap = matplotlib.cm.spectral
norm = matplotlib.colors.Normalize(vmin=200, vmax=750, clip=False)
m = matplotlib.cm.ScalarMappable(norm=norm, cmap=cmap)
wls = np.array(l.wavelengths)*1000
cols = m.to_rgba(wls)
colors = matplotlib.cycler(color=cols)
plt.gca().set_prop_cycle(colors)
m.set_array(wls)
plt.close('all')
f, axarr = plt.subplots(4, sharex=True)
f.colorbar(m)

axarr[0].plot(xs, l.combined_spline(xs)[:, :2])
axarr[0].set_title('Temperature humidity')
axarr[1].plot(xs, l.combined_spline(xs)[:, 2:3])
axarr[1].set_title('total rad')
axarr[2].plot(xs, l.combined_spline(xs)[:, 3:-1:5])

axarr[2].set_title('Spectrum')

# sp = plt.subplot(3,1,1)
# sp.plot(xs, l.combined_spline(xs)[:, 3:])

# sp = plt.subplot(3,1,2)
# sp.plot(xs, l.combined_spline(xs)[:, 2:3])

# sp = plt.subplot(3,1,3)
# sp.plot(xs, l.combined_spline(xs)[:, :2])
# plt.show()
print()
plt.show()
```
